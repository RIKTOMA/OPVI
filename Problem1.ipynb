```python
#seed=np.random.randint(100000)
#Fix the seed and the number of iterations
Each script oupts the table correspondig to the chosen parameters  in blocks, the first for the cube, the second for the sphere the thrid for the simplex
seed=60776
iters=100
sets=["cubo","sfera","simplesso"]
```

```python
#parameters for tab1
n=50
l=10
L=20

eps=0.01
rho=1
sigma=1.2


tabs={}
for j in range(3):
    parametri={}
    for i in range(4):
        
        parametri[i]=[n,l,L,sets[j],iters,eps,rho,sigma,i*5]
    tabs[j]=tables(parametri,seed)

for j in range(3):
    print(pytolatex(np.round(tabs[j],4)))
```
